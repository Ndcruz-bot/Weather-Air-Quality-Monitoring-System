```python
import matplotlib.pyplot as plt


# WHO Thresholds (example values)
WHO_PM25 = 50
WHO_PM10 = 100
WHO_NO2 = 40


class WeatherMonitoringSystem:
    def __init__(self):
        self.data = {
            "Temperature": [],
            "Humidity": [],
            "Rainfall": [],
            "PM2.5": [],
            "PM10": [],
            "NO2": []
        }

    # -----------------------------
    # Input Data
    # -----------------------------
    def enter_data(self):
        print("\n=== Enter Monthly Environmental Data ===")

        for item in self.data:
            while True:
                try:
                    value = float(input(f"Enter {item}: "))
                    self.data[item].append(value)
                    break
                except ValueError:
                    print("Invalid input! Please enter a numeric value.")

        print("\nData recorded successfully!")

    # -----------------------------
    # Statistics
    # -----------------------------
    def display_statistics(self):
        if len(self.data["Temperature"]) == 0:
            print("\nNo data available.")
            return

        print("\n========== STATISTICS ==========")

        for item, values in self.data.items():
            print(f"\n{item}")
            print(f"Minimum : {min(values):.2f}")
            print(f"Maximum : {max(values):.2f}")
            print(f"Average : {sum(values)/len(values):.2f}")

    # -----------------------------
    # Air Quality Check
    # -----------------------------
    def check_air_quality(self):
        if len(self.data["PM2.5"]) == 0:
            print("\nNo data available.")
            return

        print("\n========== AIR QUALITY REPORT ==========")

        pm25 = self.data["PM2.5"][-1]
        pm10 = self.data["PM10"][-1]
        no2 = self.data["NO2"][-1]

        if pm25 > WHO_PM25:
            print("⚠ PM2.5 exceeds WHO threshold.")
        else:
            print("✓ PM2.5 is within the safe limit.")

        if pm10 > WHO_PM10:
            print("⚠ PM10 exceeds WHO threshold.")
        else:
            print("✓ PM10 is within the safe limit.")

        if no2 > WHO_NO2:
            print("⚠ NO2 exceeds WHO threshold.")
        else:
            print("✓ NO2 is within the safe limit.")

    # -----------------------------
    # Chart
    # -----------------------------
    def generate_chart(self):
        if len(self.data["Temperature"]) == 0:
            print("\nNo data available.")
            return

        labels = list(self.data.keys())
        values = [self.data[item][-1] for item in labels]

        plt.figure(figsize=(8,5))
        plt.bar(labels, values)
        plt.title("Weather & Air Quality Data")
        plt.ylabel("Value")
        plt.xticks(rotation=20)
        plt.tight_layout()
        plt.show()

    # -----------------------------
    # Save Report
    # -----------------------------
    def save_report(self):
        if len(self.data["Temperature"]) == 0:
            print("\nNo data available.")
            return

        try:
            with open("weather_report.txt", "w") as file:
                file.write("Weather & Air Quality Report\n")
                file.write("=" * 35 + "\n")

                for item, values in self.data.items():
                    file.write(f"\n{item}\n")
                    file.write(f"Minimum : {min(values):.2f}\n")
                    file.write(f"Maximum : {max(values):.2f}\n")
                    file.write(f"Average : {sum(values)/len(values):.2f}\n")

            print("\nReport saved successfully as weather_report.txt")

        except Exception as e:
            print("Error saving file:", e)


# -----------------------------
# Main Menu
# -----------------------------
def main():

    system = WeatherMonitoringSystem()

    while True:

        print("\n========================================")
        print(" Weather & Air Quality Monitoring System")
        print("========================================")
        print("1. Enter Monthly Data")
        print("2. View Statistics")
        print("3. Check Air Quality")
        print("4. Generate Chart")
        print("5. Save Report")
        print("6. Exit")

        choice = input("\nChoose an option: ")

        if choice == "1":
            system.enter_data()

        elif choice == "2":
            system.display_statistics()

        elif choice == "3":
            system.check_air_quality()

        elif choice == "4":
            system.generate_chart()

        elif choice == "5":
            system.save_report()

        elif choice == "6":
            print("\nThank you for using the system.")
            break

        else:
            print("Invalid option. Please try again.")


if __name__ == "__main__":
    main()
```
